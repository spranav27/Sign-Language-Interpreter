```python
import cv2 as cv
import os
import time
import uuid
```

```python
images_path = "SignLanguageInterpreter/images/collectedimages"
labels = ['hi','yes','no','thankyou','iloveyou','fuckyou']
number_img = 50
```

```python
os.makedirs(images_path, exist_ok=True)
cap = cv.VideoCapture(1)
for label in labels:
    label_path = os.path.join(images_path, label)
    os.makedirs(label_path, exist_ok=True)

    print(f"Collecting images for {label}")
    time.sleep(5)

    for i in range(number_img):
        ret, frame = cap.read()
        if not ret:
            print("Camera error")
            break

        imgname = os.path.join(
            label_path,
            f"{label}.{uuid.uuid1()}.jpg"
        )

        cv.imwrite(imgname, frame)
        cv.imshow("Image Collection", frame)

        print(f"Saved {imgname}")
        time.sleep(0.2)

        if cv.waitKey(1) & 0xFF == ord('q'):
            break

cap.release()
print("Image collection completed")
```
